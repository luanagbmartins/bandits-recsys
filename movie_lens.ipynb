```python
df = ratings_df.copy()
df = df[df["rating"] >= 4]
df_list = []
for g, df_g in df.groupby("user_id"):
    list_of_indexes = []
    df_g["movie_id"].rolling(5, min_periods=1).apply(
        (lambda x: list_of_indexes.append(x.astype(int).tolist()) or 0),
        raw=False,
    )
    df_g["item_id_history"] = list_of_indexes
    df_g["item_id_history"] = df_g["item_id_history"].shift(1)

    Item_prev = (
        df_g["item_id_history"]
        .apply(lambda x: [] if x is np.nan else x)
        .apply(lambda x: ([""] * (5 - len(x)) + x)[:5])
    )
    df_list.append(Item_prev)

df["item_id_history"] = pd.concat(df_list)
```
